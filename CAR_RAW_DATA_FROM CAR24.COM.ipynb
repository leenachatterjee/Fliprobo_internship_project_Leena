```python
#Importing the library
import pandas as pd
import numpy as np
import requests
import selenium
from selenium import webdriver
from bs4 import BeautifulSoup
#importing the Web driver
dr=webdriver.Chrome("chromedriver.exe")
import time

# Importing required Exceptions which needs to handled
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException
import re
```

```python
url="https://www.cars24.com/buy-used-cars/"
```

```python
dr.get(url)
```

```python
search_btn = dr.find_element_by_xpath('//div[@class="_2bO94"]/button').click()
```

```python
url=[]
```

```python

for j in dr.find_elements_by_xpath('//div[@class="row"]/div/a'):
    url.append(j.get_attribute('href'))
       
```

```python
url
```

```python
len(url)
```

```python
his=[]

car=[]
model=[]
ownername=[]
milage=[]
fueltype=[]
emi=[]
price=[]
p_year=[]
```

```python
for i in url[:500]:
    dr.get(i)
    time.sleep(3)
    
    
    
    
    try:
     his_tag=dr.find_element_by_xpath('//strong[@class="_1-PH2"]')
     his.append(his_tag.text)
    except NoSuchElementException:
     his.append(" ")
        
        
        
    try:
     c=dr.find_element_by_xpath('//div[@class="_1l4fi"]/div/div/h2')
     car.append(c.text)
    except NoSuchElementException:
     car.append(" ")



    try:
     m=dr.find_element_by_xpath('//p[@class="cvakB"]')
     model.append(m.text)
    except NoSuchElementException:
     model.append(" ")
    
    
    try:
     o1=dr.find_element_by_xpath('//ul[@class="bVR0c"]/li[2]')
     ownername.append(o1.text)
    except NoSuchElementException:
     ownername.append(" ")
    
    try:
     m1=dr.find_element_by_xpath('//ul[@class="bVR0c"]/li')
     milage.append(m1.text)
    except NoSuchElementException:
     milage.append(" ")
    
    try:
     f1=dr.find_element_by_xpath('//ul[@class="bVR0c"]/li[3]')
     fueltype.append(f1.text)
    except NoSuchElementException:
     fueltype.append(" ")
    
    
    try:
      e1=dr.find_element_by_xpath('//div[@class="_19zc5"]/div/strong')
      emi.append(e1.text)
    except NoSuchElementException:
      emi.append(" ")
    
    
    try:
     p1=dr.find_element_by_xpath('//div[@class="_19zc5"]/div[2]/span')
     price.append(p1.text)
    except NoSuchElementException:
     price.append(" ")
    
    
    try:
      p4=dr.find_element_by_xpath('//ul[@class="_2KO13"]/li[9]')
      p_year.append(p4.text)
    except NoSuchElementException:
      p_year.append(" ")
```
