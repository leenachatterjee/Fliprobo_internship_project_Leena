# <b>Question number: 9 Write a python program to scrape fresher job listings from ‘https://internshala.com/’. It should include job title,
company name, CTC, and apply date <b>

```python
page=requests.get("https://internshala.com/fresher-jobs")
page
```

```python
soup=BeautifulSoup(page.content)
soup
```

```python
titles=soup.find_all('div', class_="heading_4_5 profile")
titles
```

```python
job_titles=[]
for i in titles:
 job_titles.append(i.text.replace('\n',''))
job_titles
```

```python
companies=soup.find_all('div'  ,class_="heading_6 company_name")
companies
```

```python
job_companies=[]
for i in companies:
 job_companies.append(i.text.strip())
job_companies
```

```python
ctc=soup.find_all('div', class_="item_body")
ctc
```

```python
job_ctc=[]
for i in ctc:
    if 'LPA' in i.text.strip():
        job_ctc.append(i.text.strip())
       
      

job_ctc
```

```python
len(job_ctc)
```

```python
len(job_ctc)
```

```python
date=soup.find_all('div', class_="item_body")
date
```

```python
job_date=[]
for i in date:
    if '21' in i.text.strip():
     job_date.append(i.text.strip())
job_date
```

```python
import pandas as pd
jobs=pd.DataFrame({})

jobs['Titles']=job_titles
jobs['Company names']=job_companies
jobs['CTC']=job_ctc
jobs['apply Date']=job_date
```

```python
jobs
```
